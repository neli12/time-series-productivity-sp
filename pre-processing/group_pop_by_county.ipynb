```python
## Import libraries
import pandas as pd
```

```python
## Load data e and select counties (mun_id) and value (valor)
dat = pd.read_csv('population_2021.csv', sep = ';')
dat_new = dat[['mun_id', 'Valor']]
dat_new
```

```python
## Group by counties and sum
dat_group = dat_new.groupby(['mun_id']).sum()
dat_group.rename(columns = {'Valor': 'POP_2021'}, inplace = True)
dat_group
```

```python
## Save new dataset
dat_group.to_csv('dat_group.csv')
```
